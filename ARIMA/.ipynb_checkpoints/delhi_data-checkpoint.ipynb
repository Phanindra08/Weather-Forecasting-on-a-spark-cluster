### Differencing the series to get the p, d and q parameters for ARIMA

```python
# create a differenced series
def difference(dataset, interval=1):
	diff = list()
	for i in range(interval, len(dataset)):
		value = dataset[i] - dataset[i - interval]
		diff.append(value)
	return diff
```

```python
diff_hum=pd.Series(difference(series_hum))
diff_hum.plot()
```

```python
plot_acf(diff_hum, lags=10)
```

### ARIMA Implementation

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
from sklearn.metrics import mean_squared_error
```

```python
#Prediction for HUMIDITY

#Split the data to test and train to mesure the accuracy of the model in model

size = int(len(series_hum) * 0.9)
train, test = series_hum[0:size], series_hum[size:len(series)]

history = [x for x in train]
predictions = list()
for t in range(len(test)):
	model = ARIMA(history, order=(1,1,2))
	model_fit = model.fit(disp=0)
	output = model_fit.forecast()
	yhat = output[0]
	predictions.append(yhat)
	obs = test.iloc[t]
	history.append(obs)
	print('predicted=%f, expected=%f' % (yhat, obs))
error = mean_squared_error(test, predictions)
print('Test MSE: %.3f' % error)
# plot
#plt.plot(test)
#plt.plot(predictions, color='red')
#plt.show()
```

```python
#Prediction for TEMPERATURE

#Split the data to test and train to mesure the accuracy of the model in model

import time
size = int(len(series_temp) * 0.9)
train, test = series_temp[0:size], series_temp[size:len(series)]

history = [x for x in train]
predictions = list()
clk=[]
clk.append(time.clock())
for t in range(len(test)):
    model = ARIMA(history, order=(2,1,2))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    obs = test.iloc[t]
    history.append(obs)
    clk.append(time.clock())
    print('predicted=%f, expected=%f' % (yhat, obs))
error = mean_squared_error(test, predictions)
print('Test MSE: %.3f' % error)
# plot
#plt.plot(test)
#plt.plot(predictions, color='red')
#plt.show()
```
